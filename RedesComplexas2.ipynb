```python
#prompt 5
#Como o grau de cada nó é sempre um número inteiro,
#coloque as barras do histograma centralizada com esses números.

import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np

# Número de nós
eNe = 10000
#eNe = 4000
eNe = 1000
# Probabilidade fixa de conexão entre nós
Pe = 0.001
#Pe = 0.0004
#Pe = 0.0001

#lambda = eNe * Pe

# eNe = 1000
# Pe = 0.2

```

```python
# Criar um grafo baseado em probabilidade fixa de conexão
def gerar_grafo_aleatorio(n, p):
    G = nx.Graph()
    G.add_nodes_from(range(n))
    for i in range(n):
        for j in range(i + 1, n):
            if random.random() < p:
                G.add_edge(i, j)
    return G

G = gerar_grafo_aleatorio(eNe, Pe)

```

```python
# Desenhar o grafo
plt.figure(figsize=(6,6))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=700, font_size=12)
plt.title(f'Grafo Aleatório com {eNe} Nós (Pe={Pe})')
plt.show()
```

```python
# Criar histograma do grau dos nós
graus = [d for n, d in G.degree()]
unique_graus = sorted(set(graus))
counts = [graus.count(g) for g in unique_graus]

plt.figure(figsize=(6,4))
plt.bar(unique_graus, counts, width=0.6, color='blue', alpha=0.7, edgecolor='black', align='center')
plt.xlabel('Grau')
plt.ylabel('Frequência')
plt.title('Histograma do Grau dos Nós')
plt.xticks(unique_graus)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Criar um grafo baseado em probabilidade fixa de conexão
def gerar_grafo_aleatorio(n, p):
    G = nx.Graph()
    G.add_nodes_from(range(n))
    for i in range(n):
        for j in range(i + 1, n):
            if random.random() < p:
                G.add_edge(i, j)
    return G

# Criar um grafo de Watts-Strogatz
def gerar_grafo_watts_strogatz(n, k, p):
    return nx.watts_strogatz_graph(n, k, p)

# Função para calcular o caminho mínimo médio sobre o maior componente conectado
def calcular_caminho_minimo_medio(G):
    if nx.is_connected(G):
        return nx.average_shortest_path_length(G)
    else:
        maior_componente = max(nx.connected_components(G), key=len)
        G_maior = G.subgraph(maior_componente).copy()
        return nx.average_shortest_path_length(G_maior)

# Função para calcular o coeficiente de aglomeração médio sobre o maior componente conectado
def calcular_coeficiente_aglomeracao_medio(G):
    maior_componente = max(nx.connected_components(G), key=len)
    G_maior = G.subgraph(maior_componente).copy()
    return nx.average_clustering(G_maior)

# Função para desenhar o grafo
def desenhar_grafo(G, titulo):
    plt.figure(figsize=(6,6))
    nx.draw_circular(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=700, font_size=12)
    plt.title(titulo)
    plt.show()

# Função para criar histograma do grau dos nós
def criar_histograma_grau(G):
    graus = [d for n, d in G.degree()]
    unique_graus = sorted(set(graus))
    counts = [graus.count(g) for g in unique_graus]

    plt.figure(figsize=(6,4))
    plt.bar(unique_graus, counts, width=0.6, color='blue', alpha=0.7, edgecolor='black', align='center')
    plt.xlabel('Grau')
    plt.ylabel('Frequência')
    plt.title('Histograma do Grau dos Nós')
    plt.xticks(unique_graus)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

```

```python
# Exibir cabeçalho
print("Lambda eNe Pe eLe Ce")

# Iterar sobre os valores de Pe
for Pe in Pe_values:
    Lambda = eNe * Pe
    G = gerar_grafo_aleatorio(eNe, Pe)
    eLe = calcular_caminho_minimo_medio(G)
    Ce = calcular_coeficiente_aglomeracao_medio(G)
    print(f"{Lambda:.2f} {eNe} {Pe:.6f} {eLe:.2f} {Ce:.6f}")

# Criar e visualizar a rede de Watts-Strogatz
G_ws = gerar_grafo_watts_strogatz(eNG, eK, ePG)
#desenhar_grafo(G_ws, "Rede de Watts-Strogatz Original")
desenhar_grafo(G_ws, "Rede de Watts-Strogatz Reconectada")
```
