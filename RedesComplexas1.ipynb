```python
#prompt0: Faça um programa em Python que crie uma rede com eNe=10 nós.
#Cada nó está conectado com todos os outros nós.
#Faça uma representação gráfica dessa rede.

import networkx as nx
import matplotlib.pyplot as plt
import random

# Número de nós
eNe = 10

# Criar um grafo completo (todos os nós conectados entre si)
G = nx.complete_graph(eNe)

# Desenhar o grafo
plt.figure(figsize=(6,6))
nx.draw(G, with_labels=True, node_color='lightblue',
        edge_color='gray', node_size=700, font_size=12)
plt.title(f'Grafo Completo com {eNe} Nós')
plt.show()

```

```python
# Criar histograma do grau dos nós
graus = [d for n, d in G.degree()]
plt.figure(figsize=(6,4))
#plt.hist(graus, bins=range(min(graus), max(graus) + 2), color='blue', alpha=0.7, edgecolor='black')
plt.hist(graus, bins=10, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Grau')
plt.ylabel('Frequência')
plt.title('Histograma do Grau dos Nós')
plt.xticks(range(min(graus), max(graus) + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
#prompt 3
#Mantenha os gráficos anteriores, mas agora faça com que cada nó esteja
#conectado a outro nó com probabilidade obtida de uma distribuição uniforme.

# Número de nós
eNe = 500

# Criar um grafo baseado em probabilidade de conexão
def gerar_grafo_aleatorio(n, p=0.5):
    G = nx.Graph()
    G.add_nodes_from(range(n))
    for i in range(n):
        for j in range(i + 1, n):
            if random.uniform(0, 1) < p:
                G.add_edge(i, j)
    return G

# Probabilidade de conexão entre nós
p_conexao = 0.5
G = gerar_grafo_aleatorio(eNe, p_conexao)

# Desenhar o grafo
plt.figure(figsize=(6,6))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=700, font_size=12)
plt.title(f'Grafo Aleatório com {eNe} Nós (p={p_conexao})')
plt.show()
```

```python
# Criar histograma do grau dos nós
graus = [d for n, d in G.degree()]
plt.figure(figsize=(6,4))
#plt.hist(graus, bins=range(min(graus), max(graus) + 2), color='blue', alpha=0.7, edgecolor='black')
plt.hist(graus, bins=15, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Grau')
plt.ylabel('Frequência')
plt.title('Histograma do Grau dos Nós')
plt.xticks(range(min(graus), max(graus) + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
